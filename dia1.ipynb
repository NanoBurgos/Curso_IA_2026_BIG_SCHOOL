# **Dia 1**

En este primer día del curso aprenderás paso a paso:

✔️ Cómo crear prompts realmente efectivos adaptados a las capacidades de ChatGPT.

✔️ Cómo diseñar tus propios GPTs personalizados para que funcionen como asistentes específicos para tu trabajo o tus proyectos.

✔️ Cómo organizar tu trabajo con Projects, la nueva forma de estructurar tareas, ideas y documentos dentro de ChatGPT.



**Formas de Usar la IA**

1. BASE - nuevo chat normal (para Tareas Puntuales)

2. PROYECTOS (para varias tareas distintas que estan relacionadas en un mismo Tema)

3. GPT´S (para tareas repetitivas)

**Un Buen PROMPT tiene**

- Contexto/Rol

- Consulta/Tarea

- Especificaciones

- Criterios de Calidad

- Formato de Respuesta
